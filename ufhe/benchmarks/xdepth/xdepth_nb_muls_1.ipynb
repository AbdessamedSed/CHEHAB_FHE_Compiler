```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plain_mod_size = 17
xdepth = 20
coeff_mod_data_level_size = 666
# plain_mod_size, xdepth, coeff_mod_data_level_size
cases = [(17, 20, 658), (17, 19, 626), (17, 18, 592), (17, 17, 562), (17, 16, 532), (17, 15, 500)]
```

```python
mul_avg_loggings = []
relin_avg_loggings = []
for case in cases:
    id = f"{case[0]}_{case[1]}_{case[2]}"
    mul_file_name = f"time_logging/without_mod_switch/mul/{id}.txt"
    mul_avg_loggings.append(np.loadtxt(mul_file_name, dtype=int).mean(axis=0).mean())
    relin_file_name = f"time_logging/without_mod_switch/relin/{id}.txt"
    relin_avg_loggings.append(np.loadtxt(relin_file_name, dtype=int).mean(axis=0).mean())
print(mul_avg_loggings)
print(relin_avg_loggings)
```

```python
xs = np.arange(50, 55, 1)
emulated_times = [[0] * xs.size for _ in range(len(cases))]
fig, ax = plt.subplots()
for i in range(len(cases)):
    for j in range(xs.size):
        emulated_times[i][j] = (mul_avg_loggings[i] + relin_avg_loggings[i]) * j
    ax.plot(xs, emulated_times[i], label=f"xdepth={20-i}")
ax.legend()
```
