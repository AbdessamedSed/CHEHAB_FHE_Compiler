```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
max_xdepth = 20

max_nb_muls = 1000
x = np.arange(max_xdepth, max_nb_muls + 1, 10)
print("x =", x)
```

```python
# xdepth = 20

# Multiplication averages per level
avg_mul_per_level_raw_20 = "652521 611997 657284 578519 572867 441786 398043 526669 408736 382880 311067 258270 209279 180842 150917 127077 102872 80334 60327 44533"
avg_mul_per_level_20 = np.array(list(map(int, avg_mul_per_level_raw_20.split())))
# Relin averages per level
avg_relin_per_level_raw_20 = "356284 292128 287368 288104 248691 214300 224064 178693 154821 157169 129613 100074 80173 68544 55149 43703 32895 23629 16438 10233"
avg_relin_per_level_20 = np.array(list(map(int, avg_relin_per_level_raw_20.split())))
# Modswitch averages per level
avg_modswitch_per_level_raw_20 = (
    "15503 14667 13754 13228 12263 11546 11933 11158 10175 9532 9586 7835 6716 6347 5298 4527 3772 2646 2171 1151"
)
avg_modswitch_per_level_20 = np.array(list(map(int, avg_modswitch_per_level_raw_20.split())))

print(avg_modswitch_per_level_20.sum())

mul_time_saved_20 = (
    avg_mul_per_level_20[0] * avg_mul_per_level_20.size - avg_mul_per_level_20.sum() - avg_modswitch_per_level_20.sum()
)
print("mutiplication time saved by the use of modswitch:", mul_time_saved_20, "μs")
print(
    "using modswitch after each multiplication is equivalent to doing only",
    20 - (mul_time_saved_20 // avg_mul_per_level_20[0]),
    "mul",
)

sequential_muls_time_20 = avg_mul_per_level_20.sum() + avg_relin_per_level_20.sum() + avg_modswitch_per_level_20.sum()
print("sequential_muls_time with xdepth=20:", sequential_muls_time_20)
```

```python
y_20 = np.empty(x.size)
y_20[0] = sequential_muls_time_20
for i in range(1, x.size):
    y_20[i] = sequential_muls_time_20 + x[i] * avg_mul_per_level_20[0]
```

```python
fig, ax = plt.subplots()
ax.plot(x, y_20)
ax.set_title("Evaluation time for different number of mul operations with xdepth=20")
ax.set_xlabel("number of mul operations")
ax.set_ylabel("Evaluation time in μs")
```
